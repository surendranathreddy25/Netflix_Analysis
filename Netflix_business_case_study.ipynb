```python
# columns with null/nan values(both categorical and contious columns)

nan_columns = [col for col in df.columns if df[col].astype(str).str.contains('nan').any()]
nan_columns
```

```python
# convert date_added to datetime format

df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
# df['date_added'] = df['date_added'].fillna(pd.NaT)
df['date_added']
```

```python
# updating null values [for catagorical with unknown {col_name}, for continous with 0]

categorical_columns = ['title', 'director', 'cast', 'country','rating','description']
continous_columns = ['duration']

for col in categorical_columns:
  df[col] = df[col].fillna(f'Unknown {col}')

for col in continous_columns:
  df[col] = df[col].fillna(0)

df
```

```python
# checking null values

df.isna().sum()
```

```python
df.info()
```

```python
# comma separated columns

comma_columns_1 = [col for col in df.columns if df[col].astype(str).str.contains(',').any()]
comma_columns_1
```

### Unnesting the columns with cells having comma separated values

```python
# title and director after un-nest

title_director = df[['title', 'director']].assign(director=df['director'].str.split(',')).explode('director', ignore_index=True)
title_director
```

```python
# title and cast after un-nest

title_cast = df[['title', 'cast']].assign(cast=df['cast'].str.split(',')).explode('cast', ignore_index=True)
title_cast
```

```python
# title and country after un-nest

title_country = df[['title', 'country']].assign(country=df['country'].str.split(',')).explode('country', ignore_index=True)
title_country
```

```python
# title and listed_in after un-nest

title_listed_in = df[['title', 'listed_in']].assign(listed_in=df['listed_in'].str.split(',')).explode('listed_in', ignore_index=True)
title_listed_in
```

```python
# merging all 4 unnested columns to form a  separate data_frame

merged_df = pd.merge(title_director, title_cast, on='title', how='inner')
merged_df = pd.merge(merged_df, title_country, on='title', how='inner')
merged_df = pd.merge(merged_df, title_listed_in, on='title', how='inner')
merged_df
```

```python
df.columns
```

```python
# remaining columns from original data_frame apart from those unnested

df_1 = df[['show_id','type','title','date_added','release_year','rating','duration','description']]
df_1
```

**Final data_frame after unnesting all comma separated values and merging them.**

```python
#final data_frame

final_df = pd.merge(df_1, merged_df, on='title', how='inner')
final_df
```

6bQ) Find After how many days the movie will be added to Netflix after the release of the movie

**Observations**:

Based on the below observations from graph


*  It takes **Average number** of days to add a movie to Netflixis **2273.14 days**


*   It takes** Maximum number** of days to add a movie to Netflix is **27483.00 days**



*   It takes **Minimum number** of days to add a movie to Netflixis **-184.00 days**(almost launch on netflix)







```python
#  Converting release year to date time (date time format)
# Assumeing release date as of jan 1st of year

final_df['release_year'] = pd.to_datetime(final_df['release_year'], format='%Y')
```

```python
# 6bQ) Find After how many days the movie will be added to Netflix after the release of the movie (you can consider the recent past data)

movies_data = final_df[final_df['type'] == 'Movie'].copy()
movies_data.loc[:, 'days_to_add'] = (movies_data['date_added'] - movies_data['release_year']).dt.days
movies_added = movies_data[['title', 'release_year', 'date_added', 'days_to_add']]
movies_added = movies_added.drop_duplicates(subset=['title'], keep='first').reset_index()
movies_added
```

```python
# plot to Find After how many days the movie will be added to Netflix after the release of the movie (you can consider the recent past data)

plt.figure(figsize=(10, 6))
sns.histplot(movies_data['days_to_add'], bins=30)
plt.title('Distribution of Days to Add Movies to Netflix After Release')
plt.xlabel('Days from Release to Netflix')
plt.ylabel('Frequency')
plt.show()
```

```python
# maximum days it takes to  add movie to netflix
max_days_to_add = movies_added['days_to_add'].max()
print(f"Maximum number of days to add a movie to Netflix: {max_days_to_add:.2f}")

# minimum days it takes to  add movie to netflix
min_days_to_add = movies_added['days_to_add'].min()
print(f"Minimum number of days to add a movie to Netflix: {min_days_to_add:.2f}")

# Avg days it takes to  add movie to netflix
avg_days_to_add = round(movies_added['days_to_add'].mean(),2)
print(f"Average number of days to add a movie to Netflix: {avg_days_to_add:.2f}")
```
